```python
# create daily cohorts 
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
```

```python
# Group by CustomerID and select the InvoiceDay value
grouping = df_transactions_small.groupby('customer_id')['t_dat'] 

# Assign a minimum InvoiceDay value to the dataset
df_transactions_small['CohortDay'] = grouping.transform('min')

# View the top 5 rows
print(df_transactions_small.head(5))
```

```python
tran_year, tran_month, tran_day = get_date_int(df_transactions_small, 't_dat')
```

```python
cohort_year, cohort_month, cohort_day = get_date_int(df_transactions_small, 'CohortDay')
```

```python
years_diff = tran_year - cohort_year

# Calculate difference in months
months_diff = tran_month - cohort_month

# Calculate difference in days
days_diff = tran_day - cohort_day

# Extract the difference in days from all previous values
df_transactions_small['CohortIndex'] = years_diff * 365 + months_diff * 30 + days_diff + 1
print(df_transactions_small.head())
```
